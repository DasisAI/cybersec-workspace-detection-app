```python
import os

nb_path = dbutils.notebook.entry_point.getDbutils().notebook().getContext().notebookPath().get()
repo_ws_root = "/".join(nb_path.split("/")[:4])
repo_fs_root = f"/Workspace{repo_ws_root}"

targets = [
    ("system.access.audit", "sandbox.audit_poc.audit"),
    ("system.query.history", "sandbox.audit_poc.history"),
    ("system.access.query_history", "sandbox.audit_poc.history"),
]

# 1) 어디에 남아있는지 먼저 찾기
hits = []
for root, _, files in os.walk(repo_fs_root):
    for fn in files:
        if not fn.endswith(".py"):
            continue
        path = os.path.join(root, fn)
        with open(path, "r", encoding="utf-8") as f:
            text = f.read()
        for old, _ in targets:
            if old in text:
                hits.append((path, old))

print("FOUND:", len(hits))
for p, old in hits[:200]:
    print(old, "->", p)

# 2) 실제 치환(원하면 주석 처리 가능)
patched = 0
for path, _ in set(hits):
    with open(path, "r", encoding="utf-8") as f:
        text = f.read()
    new = text
    for old, rep in targets:
        new = new.replace(old, rep)
    if new != text:
        with open(path, "w", encoding="utf-8") as f:
            f.write(new)
        patched += 1

print("PATCHED_FILES:", patched)
```
